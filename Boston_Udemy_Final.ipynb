## Machine Learning


```python
from sklearn.model_selection import train_test_split
from sklearn import metrics #confusion matrix and accurracy 
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression 
from sklearn.tree import DecisionTreeClassifier
lf = LogisticRegression(max_iter=100000)
clf = LinearRegression()
from sklearn.cluster import KMeans # KMeans is here because Originially I wanted to do something different however I didnt have enough time
```

```python
WebML = pd.read_csv("3.1-data-sheet-udemy-courses-web-development.csv") # Recalling our CSV, Why? I just wanted to make it easier for myself
```

```python
WebML.head()# Looking at the data again 
```

```python
WebML.isnull().sum() #finding the null values, this amount isn't going to ruin our dataset so i dropped them
```

```python
WebML = WebML.dropna() #droping null values
```

```python
WebML.isnull().sum() #checking again
```

```python
df = WebML.drop(['url', 'content_duration', 'published_timestamp'], axis=1) # dropped some columns, since to my data exploration they are not needed like url. 
```

```python
df.head()#checking my new dataframe
```

```python
X = df[["price","num_reviews"]]  ##"num_lectures","Rating" variables taken out or put in
```

```python
Y = df['num_subscribers']       ## my independent variable   
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = .2) #creating the documents I need
```

## Classification Tree

```python
my_tree = DecisionTreeClassifier().fit(X_train, Y_train)
tree_pred_train = my_tree.predict(X_train)
metrics.accuracy_score(Y_train, tree_pred_train) #this is the code used from our lecture
```

```python
pd.DataFrame(metrics.confusion_matrix(Y_train, tree_pred_train, normalize="true")) #matrix to show relationships
```

```python
tree_pred_test = my_tree.predict(X_test) 
```

```python
tree_output = pd.DataFrame(tree_pred_test,
                      index=X_test.index,
                      columns=["num_subscribers"])
tree_output.to_csv("tree_pred_final.csv")
# gives a csv file
```
